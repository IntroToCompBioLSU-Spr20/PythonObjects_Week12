# Week 12 Assignment - Simulating Dispersal with Objects

For this week's assignment, we're going to revisit the dispersal simulation that we worked on in Week 8. However, this time, we're going to use object-oriented programming. Specifically, we'll create two kinds of custom objects: `individual`s and `population`s.

Each `individual` will have 3 attributes (an id number, a x-axis position, and a y-position) and 3 methods (a constructor, a method to move its position by 1 unit, and a method to print its coordinate location as x,y).

Each `population` will have 2 attributes (a list of `individual`s in the `population` and a size for its map) and 4 methods (a constructor, a method to move each `individual` in the `population` 1 unit, a method to list the coordinate locations of all `individuals` in the `population`, and a method to display the locations of all `individual`s in the `population` as a scatter plot).

More details are provided below.

```python
# Import necessary modules
import matplotlib.pyplot as plt
import random
```

```python
# Define the individual class

class individual:
    """Class to hold information on indiviuals."""# Add a docstring
        
    def __init__(self,id="LSU_1",x=0,y=0): # Add default values for id, x, and y
        """The constructor for the individual class."""
        self.id= id# The constructor initializes the 3 attributes based on the values
        # passed as arguments to the constructor.
        self.x= x
        self.y= y
                    
    def move(self): # Add self to pass arguments
        """This function moves the indiviual one unit up, down, left or right (chosen randomly), with x and y             coordinates being modified accordingly."""
        dirs = ["u","d","l","r"]
        move = random.choice(dirs) # Randomly draws a direction - u, d, l, or r
        if move == "u":
            self.y= self.y+1 #changes y value accordingly
        elif move == "d":
            self.y= self.y-1
        elif move == "l":
            self.x= self.x-1 #changes x value accordingly
        elif move == "r":
            self.x= self.x+1
        else:
            print("Something's wrong! Direction is not u, d, l, or r.")
        # INSTRUCTIONS: 
        # Similar to the code from week 8, the code in this method should cause an
        # individual to move one unit up, down, left, or right (chosen randomly).
        # The x and y coordinates for this individual should be modified accordingly.
        
    def printLocation(self):
        """"This function prints a text description of the individual's id and location (x,y)."""
        print("The individual with id (%s) has the location (%d, %d)" % (self.id, self.x, self.y)) #grabs info from the self--individual--to print the sentence
        # INSTRUCTIONS: Add necessary arguments 
        # This method should print a text description of the 
        # individual's id and location (x,y).
```

```python
# Define the population class

class population:
    """Class to hold information on populations."""
    
    def __init__(self,mapSize=10,popSize=10): # Add default values.
        """"The constructor for the population class, taking mapSize and popSize as arguments."""
        self.individuals=[] #creates an empty list for the individuals
        for i in range(popSize): #uses the established population size to run through the loop
            self.individuals.append(individual('LSU_%d' % (i+1))) #appends each individual to the list, updating the individual's ID accordingly. Uses i+1 so that the ID starts from 1 rather than 0.
        self.mapSize= mapSize #makes mapsize an attribute
        # Add a docstring
        # INSTRUCTIONS: This constructor should create a list of individuals in the population. The 
        # number of individuals should be equal to popSize and each individual should have
        # a unique id. The constructor should also record the mapSize as an attribute.

        
    def move(self): # Add necessary arguments
        """This method will move each individual in the population class by one unit."""
        for individual in self.individuals: #loops through the above-created list
            individual.move() #calls upon the move function

        # INSTRUCTIONS: 
        # Add a docstring
        # The code in this method should use a for loop and cause each individual to 
        # move one unit up, down, left, or right. This code should call the .move() method
        # of each individual.

        
    def listLocations(self): # Add necessary arguments
        # Add a docstring
        """This method will print a text description of each individual's location."""
        for individual in self.individuals: #loop through the list
            individual.printLocation() #call upon the printLocation method
        
        # INSTRUCTIONS: The code in this method should use a for loop to print a text description of
        # each individual's location. This code should call the .printLocation() method
        # of each individual.

        
    def draw(self): # Add necessary arguments
        """This method will create a scatterplot to show the locations of all individuals in the population."""
        plt.figure(figsize=(self.mapSize,self.mapSize)) #this establishes the plotsize using mapsize
        for individual in self.individuals: #need to loop through the individuals
            plt.plot(individual.x, individual.y,'bo', markersize=3) #to then plot each x and y value belonging to the indivduals. Make sure to add info about the points to make it a scatterplot. Remember that individuals is a LIST OF OBJECTS, so using indexing could not work.
        plt.show() #shows the relevant plot

        # INSTRUCTIONS: The code in this method should use the .plot() method from pyplot to 
        # create a scatterplot showing the locations of all individuals in this
        # population. The length of the x- and y-axes should each be mapSize.
```

Now that we've created the individual and population classes, let's use them to create populations of individuals and watch them disperse across a map.

```python
# Create a new population with default values
newpop=population()

# Assign the number of time steps that individuals will move
steps=20

# Use a for loop to simulate the movement of individuals for the specified number of steps.
for x in range(steps):
    newpop.move() #calls upon the move method of population class

# Use the .draw() method of the population to draw the map.
newpop.draw()

# Use the .listLocations() method to list the locations of all individuals in the population.
newpop.listLocations()
```
