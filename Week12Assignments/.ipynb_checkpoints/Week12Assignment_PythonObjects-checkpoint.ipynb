# Week 12 Assignment - Simulating Dispersal with Objects

For this week's assignment, we're going to revisit the dispersal simulation that we worked on in Week 8. However, this time, we're going to use object-oriented programming. Specifically, we'll create two kinds of custom objects: `individual`s and `population`s.

Each `individual` will have 3 attributes (an id number, a x-axis position, and a y-position) and 3 methods (a constructor, a method to move its position by 1 unit, and a method to print its coordinate location as x,y).

Each `population` will have 2 attributes (a list of `individual`s in the `population` and a size for its map) and 4 methods (a constructor, a method to move each `individual` in the `population` 1 unit, a method to list the coordinate locations of all `individuals` in the `population`, and a method to display the locations of all `individual`s in the `population` as a scatter plot).

More details are provided below.

```python
# Import necessary modules
from matplotlib import pyplot as plt # for scatter plot
import random  # for random.choice
```

```python
# Define the individual class
class individual:
    """Stores the id, coordinates, and methods of each individual"""

    def __init__(self,id=None,x=0,y=0): 
        """Initializes the ID and coordinates of each individual"""
        
        self.id = id
        self.x = x
        self.y = y

    def move(self, mapSize):
        """Randomly moves each individual by one step in a random direction within the borders of the map"""
        
        directions = ["up", "down", "left", "right"]
        choice = random.choice(directions)
        
        if choice == "up" and self.y < mapSize:
            self.y += 1
        elif choice == "down" and self.y > 0:
            self.y -= 1
        elif choice == "left" and self.x > 0:
            self.x -= 1
        elif choice == "right" and self.x < mapSize:
            self.x += 1
     
    def printLocation(self): 
        """Prints the ID and location in coordinates of each individual"""
        print("Individual " + str(self.id) + ":(" + str(self.x) + "," + str(self.y) + ")")
```

```python
# Define the population class
class population:
    """Contains map and population size as well as methods to control multiple individuals at once"""
    
    def __init__(self,mapSize=100,popSize=100): # Add default values.
        """Initializes map size and population size. Creates a population of individuals starting in center of map"""
        
        self.mapSize = mapSize
        self.popSize = popSize
        
        centerX = centerY = int(self.mapSize / 2)  # Center of Map
        self.indList = []  # contains list of individuals
        
        for i in range(popSize):
            ind = individual(i + 1, centerX, centerY)  # creates individual and adds it to population
            self.indList.append(ind)
      
    def move(self): 
        """Runs move() for each individual"""
        
        for i in self.indList:
            i.move(self.mapSize)
      
    def listLocations(self): 
        """Runs printLocation() for each individual"""
        
        for i in self.indList:
            i.printLocation()
       
    def draw(self, axisRange=None): 
        """Draws a map with the positions of each individual using pyplot"""
        
        if axisRange == None:  # set the range of the x and y axis
            axisRange = self.mapSize
    
        x = []  # list that contain all coordinates 
        y = []
        
        for i in self.indList:
            x.append(i.x)
            y.append(i.y)
            
        plt.ylim(0, axisRange)  # change axis ranges of map
        plt.xlim(0, axisRange)
        plt.scatter(x, y)  # use scatter plot instead of line plot
        plt.show()
        
```

Now that we've created the individual and population classes, let's use them to create populations of individuals and watch them disperse across a map.

```python
pop = population()  # initialize first population with default values for map and pop size
steps = 1000  # number of moves each individual will make

for _ in range(steps):
    pop.move()
    
pop.draw()  
pop.listLocations()
```

```python
pop2 = population(50, 200)
steps = 5000

for _ in range(steps):
    pop2.move()
    
pop2.draw(100)  # change axis range of map to be able to compare densities of the two populations
pop2.listLocations()
```
